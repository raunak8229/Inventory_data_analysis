## ABC Analysis (Inventory Classification)

### Prepare ABC base table

#### Aggregate sales by product

```python
abc_df = (sales.groupby('description')['total_sales_amount'].sum().reset_index())
```

```python
abc_df = abc_df.sort_values(by='total_sales_amount', ascending=False)
```

```python
abc_df.head(10)
```

### Calculate cumulative contribution

#### Total sales value

```python
total_sales_value = abc_df['total_sales_amount'].sum()
```

```python
total_sales_value
```

#### Percentage contribution

```python
abc_df['value_pct'] = (abc_df['total_sales_amount'] / total_sales_value).round(3)
```

```python
abc_df.head(10)
```

```python
abc_df['cum_value_pct'] = abc_df['value_pct'].cumsum()
```

```python
abc_df.head(10)
```

```python
def abc_classification(cum_pct):
    if cum_pct <= 0.70:
        return 'A'
    elif cum_pct <= 0.90:
        return 'B'
    else:
        return 'C'

abc_df['abc_category'] = abc_df['cum_value_pct'].apply(abc_classification)
```

```python
abc_df['abc_category'].value_counts()
```

```python
abc_df.head(10)
```

```python
abc_df['cum_value_pct'].plot(
    figsize=(12,4),
    title='ABC Analysis  Cumulative Sales Value'
)
```

### Prepare demand data

#### Annual demand per product

```python
annual_demand = (
    sales
    .groupby('description')['salesquantity']
    .sum()
    .reset_index()
    .rename(columns={'salesquantity': 'annual_demand'})
)
```

```python
annual_demand
```

**Ordering cost per order = 100(currency units)**

```python
ORDERING_COST = 100
```

#### Average purchase price

```python
avg_purchase_price = (
    pp
    .groupby('description')['purchaseprice']
    .mean()
    .reset_index()
)
```

```python
avg_purchase_price
```

#### Merge with demand

```python
eoq_df = pd.merge(
    annual_demand,
    avg_purchase_price,
    on='description',
    how='left'
)
```

```python
eoq_df
```

#### Calculate holding cost

```python
HOLDING_RATE = 0.20
eoq_df['holding_cost'] = eoq_df['purchaseprice'] * HOLDING_RATE
```

```python
eoq_df
```

### Claculate EOQ(Economic Order Quantity)

```python
eoq_df['eoq'] = np.sqrt(
    (2 * eoq_df['annual_demand'] * ORDERING_COST) / eoq_df['holding_cost'])
```

```python
eoq_df['eoq'] = eoq_df['eoq'].round(0)
```

```python
eoq_df
```

#### Handle edge cases

Some products may have:
- Zero demand
- Missing purchase price

**Fix this:**

```python
eoq_df = eoq_df.dropna(subset=['holding_cost'])
eoq_df = eoq_df[eoq_df['annual_demand'] > 0]
```

```python
eoq_df
```

#### Review EOQ result

```python
eoq_df[['description', 'annual_demand', 'purchaseprice', 'holding_cost', 'eoq']].head(10)
```

#### Merge EOQ with ABC results:

```python
eoq_df = pd.merge(
    eoq_df,
    abc_df[['description', 'abc_category']],
    on='description',
    how='left'
)
```

```python
eoq_df.head(10)
```

## Reorder Point (ROP) Analysis

EOQ told us **how much** to order,
ROP tells us **when** to order

### Calculate average daily demand

#### Total order period(days)

```python
# Total sales period (days)
sales_days = (sales['salesdate'].max() - sales['salesdate'].min()).days
sales_days
```

#### Average daily demand per product

```python
daily_demand = (
    sales
    .groupby('description')['salesquantity']
    .sum()
    .reset_index()
)

daily_demand['avg_daily_demand'] = daily_demand['salesquantity'] / sales_days
daily_demand = daily_demand[['description', 'avg_daily_demand']]
```

```python
daily_demand
```

#### Assume lead time

```python
LEAD_TIME = 10
```

#### DAILY DEMAND PER PRODUCT

```python
daily_sales = (
    sales
    .groupby(['description', 'salesdate'])['salesquantity']
    .sum()
    .reset_index()
)
```

```python
daily_sales
```

#### Standard deviation of daily demand

```python
demand_std = (
    daily_sales
    .groupby('description')['salesquantity']
    .std()
    .reset_index()
    .rename(columns={'salesquantity': 'demand_std'})
)
```

```python
demand_std
```

### Claculate Safety stock

**Safety Stock=ZL**

Where:

Z = service level factor

 = demand standard deviation

L = lead time

**Well use 95% service level, Z value = 1.65**  

```python
Z = 1.65
```

```python
safety_stock = demand_std.copy()
safety_stock['safety_stock'] = Z * safety_stock['demand_std'] * np.sqrt(LEAD_TIME)
```

#### Calculate Reorder Point (ROP)

**Merge All Components**

```python
rop_df = daily_demand.merge(safety_stock, on='description', how='left')
```

***Final ROP formula***

```python
rop_df['rop'] = (
    (rop_df['avg_daily_demand'] * LEAD_TIME) +
    rop_df['safety_stock']
)
```

**Round for practicality**

```python
rop_df['rop'] = rop_df['rop'].round(0)
rop_df['safety_stock'] = rop_df['safety_stock'].round(0)
```

### Combine ROP with EOQ & ABC (FINAL MASTER TABLE)

```python
final_inventory_policy = (
    eoq_df
    .merge(rop_df[['description', 'rop', 'safety_stock']], on='description', how='left')
)
```

```python
final_inventory_policy
```

#### Business interpretation :-

Reorder point ensures stock availability during lead time

Safety stock protects against demand uncertainty

Higher service levels increase safety stock

ABC items justify different service levels

## Key Insights

#### **Insight 1:** Sales follow the Pareto principle

A small percentage of products contribute to the majority of revenue

ABC analysis showed A-category items generate ~70% of sales value

 ***Implication:*** Not all products should be managed equally

#### **Insight 2:** Presence of slow-moving products

Many products have very low annual demand

These products fall mostly into C category

 ***Implication:*** Risk of dead stock and tied-up capital

#### **Insight 3:** High dependency on a few vendors

A limited number of vendors contribute most sales value

Vendor concentration risk exists

 ***Implication:*** Supply disruptions could heavily impact revenue

#### **Insight 4:** Inventory levels not aligned with demand

Ending inventory differs significantly from beginning inventory

Some low-demand items still hold high inventory

 ***Implication:*** Overstocking and inefficient inventory allocation

#### **Insight 5:** EOQ and ROP reveal inefficiencies

EOQ values differ significantly from current order behavior

Reorder points with safety stock reduce stockout risk

 ***Implication:*** Current ordering likely causes unnecessary holding or stockouts

### Strategic Recommendations

####  A-Category Products

**Recommendations:**

- Apply strict EOQ-based ordering

- Maintain higher safety stock

- Monitor inventory frequently

- Use higher service levels (95%)

**Business impact:**

 Reduced stockouts

 Stable revenue flow

####  B-Category Products

**Recommendations:**

- Moderate EOQ adherence

- Monthly or bi-monthly review

- Balanced service level

**Business impact:**

 Controlled cost without heavy monitoring

####  C-Category Products

**Recommendations:**

- Simplify ordering (bulk or periodic)

- Reduce safety stock

- Identify candidates for discontinuation

**Business impact:**
                      
 Reduced holding cost
                      
 Lower capital lock-in

### Vendor & Procurement Recommendations

- Renegotiate pricing with high-cost vendors

- Reduce dependency on single vendors for A-items

- Consolidate vendors for C-items to reduce admin cost
